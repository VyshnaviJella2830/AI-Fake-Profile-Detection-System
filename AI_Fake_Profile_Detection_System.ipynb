```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# ðŸ”¹ Sample Dataset (You can later replace with real dataset)
data = {
    "followers": [50, 5000, 30, 10000, 80, 2000, 45, 7000],
    "following": [1000, 300, 900, 200, 850, 400, 950, 350],
    "posts": [5, 300, 2, 500, 3, 150, 1, 400],
    "bio_length": [10, 120, 5, 150, 8, 90, 4, 130],
    "is_fake": [1, 0, 1, 0, 1, 0, 1, 0]  # 1 = Fake, 0 = Real
}

df = pd.DataFrame(data)

# Features & Target
X = df[["followers", "following", "posts", "bio_length"]]
y = df["is_fake"]

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Accuracy
y_pred = model.predict(X_test)
print("ðŸ”¥ Model Accuracy:", round(accuracy_score(y_test, y_pred)*100,2), "%")

# ðŸ”¹ User Input Prediction
print("\nEnter Profile Details:")

followers = int(input("Followers count: "))
following = int(input("Following count: "))
posts = int(input("Number of posts: "))
bio_length = int(input("Bio length (characters): "))

prediction = model.predict([[followers, following, posts, bio_length]])

if prediction[0] == 1:
    print("\nðŸš¨ This profile is likely FAKE")
else:
    print("\nâœ… This profile is likely REAL")
```
